### 分析

在搜尋框內可輸入

1. 特定粉絲首頁：查詢粉專網址後貼上

    【特定粉專首頁】
    如今天如果我想搜尋＂蔡英文＂總統的粉專，可以用臉書查詢該粉專首頁之後，將網址貼在搜尋欄內，就會出現相對應的粉專
    ![title](img/1.jpg)
    https://www.facebook.com/tsaiingwen/?ref=br_rs
    
    另外一次可以輸入多個候選人進行查詢，在選取蔡英文的粉專後，也可以加上韓國瑜的粉專
    ![title](img/2.jpg)
    https://www.facebook.com/twherohan/
    
     
2. 自行輸入關鍵字
    搜尋欄內可以自訂關鍵字，系統會抓出包含關鍵字的文章
    要特別注意的是，同一個人物或議題可能會有很多不同的稱呼方式
    比如說"蔡英文"，就可能會有"蔡總統"、"小英"、"蔡英文"等等的稱呼
    在搜尋欄中，也有類似程式語言的語法，可以善用'or'和'and'的概念
    
    例如：
    
    。or 或 (or和字之間要空格)
    和蔡英文有關的貼文：
    蔡總統 or 小英 or 蔡英文 
    
    。and 且 ( and和字之間要空格，也可以省去 and 不寫 )
    和蔡英文有關，且提到「兩岸」關鍵字的新聞：
    > 蔡總統 兩岸 
    > 蔡總統 and 兩岸 
    
    。not 排除 (not和字之間要空格)
    和蔡英文有關，但排除有「選舉」關鍵字的貼文：
    > 蔡英文 - 選舉
    > 蔡英文 not 選舉
    
    。可以混合一起用
    和蔡英文相關且提到肺炎的貼文：
    > (蔡總統 or 小英 or 蔡英文) 肺炎

3. 特定子集資料庫

    。seg:KOL 意見領袖資料庫
    
    目前聲量高的意見領袖有哪些
    > seg:KOL
    
    意見領袖對特定議題/人物的看法
    > seg:KOL 空污
    
    。seg:media 新聞媒體資料庫
    
    目前聲量高的新聞媒體有哪些
    > seg:media
    
    新聞媒體對特定議題/人物的看法
    > seg:media 肺炎
    
    。name: 抓有特定名字的粉絲頁
    
    國民黨的同溫層有哪些
    > name:國民黨
